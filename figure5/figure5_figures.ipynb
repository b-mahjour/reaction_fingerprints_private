```python
import pandas as pd
import os

files = sorted(os.listdir())
```

```python

ofiles = []
for k in files:
    if "ipynb" in k:
        continue
    if k[0] == ".":
        continue
    if "tsne" in k:
        continue
    if ".ai" in k:
        continue
    print(k)
    ofiles.append(k)
    
    
```

```python
import numpy as np
from rdkit import Chem
from rdkit.Chem import DataStructs, AllChem, MACCSkeys
from rdkit import RDLogger
import matplotlib.pyplot as plt
RDLogger.DisableLog('rdApp.*')        

data = pd.read_csv(ofiles[0])

headers = ["Nucleophile SMILES", "Electrophile SMILES", "Catalyst1 SMILES", "Catalyst2 SMILES", "Ligand1 SMILES", "Ligand2 SMILES", \
           "Catalyst2 SMILES", "BaseAcid SMILES", "ReductantOxidant SMILES", "Additive SMILES", "Solvent1 SMILES"]

rfps = []
output = []
for i,k in data.iterrows():
    rfp = np.zeros(2048)
    for j in headers:
        try:
            mol = Chem.MolFromSmiles(k[j])
            fp = np.zeros((1,))
            DataStructs.ConvertToNumpyArray(AllChem.GetMorganFingerprintAsBitVect(mol, 4), fp)
            n = 1
            if "Electrophile" in j:
                n = 5
            rfp = rfp + fp*n
        except:
            continue
    output.append(k["Output Value"])
    rfps.append(rfp)
print(len(rfps))

from sklearn.manifold import TSNE

X_TSNE_RFP = TSNE(n_components=2, n_jobs=-1, perplexity=10, random_state=1).fit_transform(rfps)

fig, ax = plt.subplots(figsize=(1.5,1.5), dpi=300)
ax.scatter(X_TSNE_RFP[:,0], X_TSNE_RFP[:,1], cmap='viridis', c=output, alpha=1, marker='.',s=10)
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel("tsne$_x$", fontsize=6, fontfamily="arial")
ax.set_ylabel("tsne$_y$", fontsize=6, fontfamily="arial")

plt.subplots_adjust(wspace=0, hspace=0)
plt.savefig('tsneA.png', dpi=300, bbox_inches='tight',pad_inches=0.01)
```

```python
import numpy as np
from rdkit import Chem
from rdkit.Chem import DataStructs, AllChem, MACCSkeys
from rdkit import RDLogger
import matplotlib.pyplot as plt
RDLogger.DisableLog('rdApp.*')        

print(ofiles[7])
data = pd.read_excel(ofiles[7])

headers = ["Nucleophile SMILES", "Electrophile SMILES", "Catalyst1 SMILES", "Catalyst2 SMILES", "Ligand1 SMILES", "Ligand2 SMILES", \
           "Catalyst2 SMILES", "BaseAcid SMILES", "ReductantOxidant SMILES", "Additive SMILES"]




ohe = {}
rfps = []
ohefps = []
output = []
for i,k in data.iterrows():
    rfp = np.zeros(2048)
    ohefp = []
    for j in headers:
        try:
            mol = Chem.MolFromSmiles(k[j])
            fp = np.zeros((1,))
            DataStructs.ConvertToNumpyArray(AllChem.GetMorganFingerprintAsBitVect(mol, 4), fp)
            n = 1
            if "ReductantOxidant" in j:
                n = 3
            rfp = rfp + fp*n
            if j not in ohe:
                ohe[j] = []
            
            sm = Chem.CanonSmiles(k[j])
            ohe[j].append(sm)
            
            ohefp.append(ohe[j].index(sm))
        except:
            continue
    output.append(k["Output Value"])
    rfps.append(rfp)
    ohefps.append(ohefp)
```
